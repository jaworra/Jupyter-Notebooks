```python
# If there isn't 6 in count, remove from analysis
mask = merged_inner['group_by_counts'] == 6
here_links_all_sites_new = merged_inner.loc[mask]
here_links_all_sites_new.sort_values(by=['index']) # sort on previous value
```

#### Removed 1932 records (24846 - 22914).

```python
#Calculate travael time in seconds
here_links_all_sites_new['Travel_Time_Sec']= (here_links_all_sites_new['LENGTH']/1000)/here_links_all_sites_new['MEAN']*3600
here_links_all_sites_new
```

```python
#Calcuate the travel times and group by time (Group by date/time)
here_links_all_sites_new_group_time =here_links_all_sites_new.groupby(['DATE-TIME'])['DATE-TIME','Travel_Time_Sec'].sum()
here_links_all_sites_new_group_time
```

```python
#Pull out hour into new column
here_links_all_sites_new_group_time['Date-Time'] = here_links_all_sites_new_group_time.index
here_links_all_sites_new_group_time['Date-Time'] = pd.to_datetime(here_links_all_sites_new_group_time['Date-Time'])
here_links_all_sites_new_group_time['Hour'] = pd.to_datetime(here_links_all_sites_new_group_time['Date-Time']).dt.hour
here_links_all_sites_new_group_time
```

```python
#new dataframe with statistics calculations
here_links_load_by_hour_avg_statistic =here_links_all_sites_new_group_time.groupby(['Hour'])['Date-Time','Travel_Time_Sec'].mean()
here_links_load_by_hour_avg_statistic['Min']=here_links_all_sites_new_group_time.groupby(['Hour'])['Travel_Time_Sec'].min()
here_links_load_by_hour_avg_statistic['Max']=here_links_all_sites_new_group_time.groupby(['Hour'])['Travel_Time_Sec'].max()
here_links_load_by_hour_avg_statistic['Median']=here_links_all_sites_new_group_time.groupby(['Hour'])['Travel_Time_Sec'].median()

#add a column with counts to give statistics a reference
count_df = here_links_all_sites_new_group_time.groupby(['Hour']).size().reset_index(name='Counts')
here_links_load_by_hour_avg_statistic = pd.merge(left=count_df,right=here_links_load_by_hour_avg_statistic, left_on='Hour', right_on='Hour')

#round values
here_links_load_by_hour_avg_statistic.round({"Travel_Time_Sec":1, "Min":1, "Max":1, "Median":1}) 
```

```python
here_links_load_by_hour_avg_statistic[['Min','Max','Travel_Time_Sec','Median']].plot(figsize=(15,10), grid= True);
```
